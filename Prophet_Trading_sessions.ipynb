```python
from prophet import Prophet  # Use 'prophet' instead of 'fbprophet'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Load the datasets
stock_data = pd.read_csv('stock_yfinance_data.csv')
sentiment_data = pd.read_csv('stock_tweets.csv')

# Preprocess sentiment data
analyzer = SentimentIntensityAnalyzer()
sentiment_data['sentiment_score'] = sentiment_data['Tweet'].apply(lambda x: analyzer.polarity_scores(x)['compound'])

# Convert Date columns to datetime without timezone information
stock_data['Date'] = pd.to_datetime(stock_data['Date']).dt.tz_localize(None)
sentiment_data['Date'] = pd.to_datetime(sentiment_data['Date']).dt.tz_localize(None)

# Average sentiment score per stock per day
average_sentiment = sentiment_data.groupby(['Stock Name', 'Date'])['sentiment_score'].mean().reset_index()

# Merge finance data with sentiment data
merged_data = pd.merge(stock_data, average_sentiment, on=['Stock Name', 'Date'], how='left')
merged_data['sentiment_score'].fillna(0, inplace=True)

# Unique stocks list
unique_stocks = merged_data['Stock Name'].unique()

# Initialize portfolio variables
initial_investment = 10000
portfolio_value = initial_investment
equity_curve = [initial_investment]
trades = []  # To store information about each trade (buy/sell)

# Loop through each stock
for stock in unique_stocks:
    print(f"Processing stock: {stock}")
    stock_data = merged_data[merged_data['Stock Name'] == stock].copy()

    # Prepare data for Prophet
    stock_data['ds'] = stock_data['Date']
    stock_data['y'] = stock_data['Close']

    # Initialize the Prophet model with sentiment regressor
    prophet_with_sentiment = Prophet(daily_seasonality=True, yearly_seasonality=True, weekly_seasonality=True)
    prophet_with_sentiment.add_regressor('sentiment_score')
    prophet_with_sentiment.fit(stock_data[['ds', 'y', 'sentiment_score']])

    # Prepare future dataframe
    future_dates = prophet_with_sentiment.make_future_dataframe(periods=365)
    future_sentiment = pd.merge(future_dates, stock_data[['ds', 'sentiment_score']], on='ds', how='left')
    future_sentiment['sentiment_score'].fillna(0, inplace=True)

    # Predict future stock prices with sentiment
    forecast = prophet_with_sentiment.predict(future_sentiment)

    # Extract predicted prices
    predicted_prices = forecast[['ds', 'yhat']]

    # Generate trading signals based on predicted price change
    actual = stock_data['y'].values
    predicted = forecast['yhat'][:len(actual)].values
    min_length = min(len(actual), len(predicted))
    signals = []
    threshold = 0.01  # Reduced threshold for increased sensitivity
    for i in range(1, min_length):
        price_change = (predicted[i] - actual[i - 1]) / actual[i - 1]
        if price_change > threshold:
            signals.append(1)  # Buy
            trades.append({'Date': stock_data['ds'].values[i], 'Stock': stock, 'Action': 'Buy', 'Price': actual[i]})
        elif price_change < -threshold:
            signals.append(-1)  # Sell
            trades.append({'Date': stock_data['ds'].values[i], 'Stock': stock, 'Action': 'Sell', 'Price': actual[i]})
        else:
            signals.append(0)  # Hold

    # Calculate equity curve for each signal
    portfolio_value = initial_investment  # Reset to initial investment for each stock
    equity_curve_stock = [portfolio_value]  # Start equity curve from initial investment
    for i in range(1, len(signals)):
        price_change = (predicted[i] - actual[i - 1]) / actual[i - 1]
        if signals[i - 1] == 1:  # Buy
            portfolio_value *= (1 + price_change)
        elif signals[i - 1] == -1:  # Sell
            portfolio_value *= (1 - price_change)
        equity_curve_stock.append(portfolio_value)
    equity_curve.extend(equity_curve_stock[1:])  # Append to the main equity curve, skipping duplicate initial value

# Plot the equity curve for the entire trading strategy
plt.figure(figsize=(12, 8))
plt.plot(range(len(equity_curve)), equity_curve, label='Equity Curve', color='green')
plt.xlabel('Trades (Chronological)')
plt.ylabel('Portfolio Value ($)')
plt.title('Equity Curve for Trading Strategy')
plt.legend()
plt.tight_layout()
plt.show()

# Print summary of trades
print("\nSummary of Trades:")
for trade in trades:
    print(f"Date: {trade['Date']}, Stock: {trade['Stock']}, Action: {trade['Action']}, Price: {trade['Price']}")

# Calculate the final value of the equity curve
final_portfolio_value = equity_curve[-1]

# Calculate the total profit or loss
total_profit = final_portfolio_value - initial_investment

# Calculate the Return on Investment (ROI)
roi = (total_profit / initial_investment) * 100

# Print the results
print(f"\nFinal Portfolio Value: ${final_portfolio_value:.2f}")
print(f"Total Profit: ${total_profit:.2f}")
print(f"Return on Investment (ROI): {roi:.2f}%")

# Determine profitability
if total_profit > 0:
    print("The strategy is profitable.")
else:
    print("The strategy is not profitable.")
```
