```python
from prophet import Prophet  # Use 'prophet' instead of 'fbprophet'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Load the datasets
stock_data = pd.read_csv('stock_yfinance_data.csv')
sentiment_data = pd.read_csv('stock_tweets.csv')

# Preprocess sentiment data
analyzer = SentimentIntensityAnalyzer()
sentiment_data['sentiment_score'] = sentiment_data['Tweet'].apply(lambda x: analyzer.polarity_scores(x)['compound'])

# Convert Date columns to datetime without timezone information
stock_data['Date'] = pd.to_datetime(stock_data['Date']).dt.tz_localize(None)
sentiment_data['Date'] = pd.to_datetime(sentiment_data['Date']).dt.tz_localize(None)

# Average sentiment score per stock per day
average_sentiment = sentiment_data.groupby(['Stock Name', 'Date'])['sentiment_score'].mean().reset_index()

# Merge finance data with sentiment data
merged_data = pd.merge(stock_data, average_sentiment, on=['Stock Name', 'Date'], how='left')
merged_data['sentiment_score'].fillna(0, inplace=True)

# Unique stocks list
unique_stocks = merged_data['Stock Name'].unique()

# Loop through each stock
for stock in unique_stocks:
    print(f"Processing stock: {stock}")
    stock_data = merged_data[merged_data['Stock Name'] == stock].copy()

    # Prepare data for Prophet
    stock_data['ds'] = stock_data['Date']
    stock_data['y'] = stock_data['Close']

    # Initialize the Prophet model
    prophet = Prophet(daily_seasonality=True, yearly_seasonality=True, weekly_seasonality=True)

    # Add sentiment score as a regressor
    prophet.add_regressor('sentiment_score')

    # Fit the model
    prophet.fit(stock_data[['ds', 'y', 'sentiment_score']])

    # Prepare future dataframe
    future_dates = prophet.make_future_dataframe(periods=365)
    future_sentiment = pd.merge(future_dates, stock_data[['ds', 'sentiment_score']], on='ds', how='left')
    future_sentiment['sentiment_score'].fillna(0, inplace=True)

    # Predict future stock prices
    forecast = prophet.predict(future_sentiment)

    # Extract important fields from forecast
    predicted_prices = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]

    # Evaluate model accuracy (using MAE and RMSE)
    actual = stock_data['y'].values
    predicted = forecast['yhat'][:len(actual)].values
    min_length = min(len(actual), len(predicted))
    mae = mean_absolute_error(actual[:min_length], predicted[:min_length])
    rmse = np.sqrt(mean_squared_error(actual[:min_length], predicted[:min_length]))
    print(f'Mean Absolute Error (MAE) for {stock}: {mae}')
    print(f'Root Mean Squared Error (RMSE) for {stock}: {rmse}')

    # Generate trading signals based on predicted price change
    signals = []
    signal_dates = []
    threshold = 0.01  # Reduced threshold for increased sensitivity
    for i in range(1, min_length):
        price_change = (predicted[i] - actual[i - 1]) / actual[i - 1]
        if price_change > threshold:
            signals.append(1)  # Buy
        elif price_change < -threshold:
            signals.append(-1)  # Sell
        else:
            signals.append(0)  # Hold
        signal_dates.append(stock_data['ds'].values[i])

    # Print trading signals summary
    print(f"Trading signals for {stock}: {signals[:10]} (showing first 10 signals)")

    # Visualize actual vs predicted price with trading signals
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data['ds'][:min_length], actual[:min_length], label='Actual Price', color='blue')
    plt.plot(stock_data['ds'][:min_length], predicted[:min_length], label='Predicted Price', color='red', linestyle='--')

    # Plot trading signals
    for i, signal in enumerate(signals):
        if signal == 1:
            plt.scatter(signal_dates[i], actual[i], marker='^', color='green', s=100, label='Buy Signal' if not any(s == 1 for s in signals[:i]) else "")
        elif signal == -1:
            plt.scatter(signal_dates[i], actual[i], marker='v', color='red', s=100, label='Sell Signal' if not any(s == -1 for s in signals[:i]) else "")

    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(f'Actual vs Predicted Price with Trading Signals for {stock}')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Save forecast to CSV
    predicted_prices.to_csv(f'prophet_forecast_{stock}.csv', index=False)
```
